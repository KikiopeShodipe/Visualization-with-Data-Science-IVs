```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
#create the Blood pressure readings
bp_readings = [120, 130, 140, 150, 160, 145, 155, 165, 165, 175]
#create a dataframe
df = pd.DataFrame({'Blood Pressure': bp_readings})
df
```

```python
sns.histplot(bp_readings, bins=6, kde=True, color="red")
plt.title("Distribution of High Blood Pressure Readings")
plt.xlabel("Blood Pressure")
plt.ylabel('count')
plt.show()
```

```python
#Bimodal histogram
import numpy as np
data1=np.random.normal(0,1,150)
data2=np.random.normal(10,2,150)
data=np.concatenate((data1,data2))
#histogram
plt.hist(data, bins=10, color='violet', edgecolor='black')
plt.title("Bimodal Histogram")
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")
plt.show()
```

```python
#positively skewed-increasing upwards
data=np.random.exponential(scale=1, size=100) + 120
plt.hist(data, bins=10, color='orange', edgecolor='black')
plt.title("Positively Skewed Histogram")
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")
plt.show()
```

```python
#positively skewed-increasing upwards
data = -np.random.exponential(scale=1, size=100) + 180
plt.hist(data, bins=10, color='yellow', edgecolor='black')
plt.title("Right-Skewed Histogram")
plt.xlabel("Blood Pressure")
plt.ylabel("Frequency")
plt.show()
```
